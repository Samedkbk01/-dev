```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

bgr = cv2.imread("para.jpg")
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
```

```python
gri_para = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(gri_para, cmap='gray')
plt.axis("off")
```

```python
bulanik_para = cv2.GaussianBlur(rgb, (5, 5), 2)
plt.imshow(bulanik_para)
plt.axis("off")
```

```python
_, binary_para = cv2.threshold(gri_para, 138, 217, cv2.THRESH_BINARY)
plt.imshow(binary_para, cmap='gray')
plt.axis("off")
```

```python
kenar_para = cv2.Canny(gri_para, 130, 245)
plt.imshow(kenar_para, cmap='gray')
plt.axis("off")
```

```python
bgr = cv2.imread("para.jpg")
kenar_para = cv2.dilate(kenar_para, (3,3), iterations=80)
contours, hierarchy = cv2.findContours(kenar_para, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
kontorlu_para = cv2.drawContours(rgb, contours, -1, (255,0,0), 4)

for i, kontor in enumerate(contours):
  if cv2.contourArea(kontor) > 180:
    x, y, w, h = cv2.boundingRect(kontor)
    cv2.rectangle(kontorlu_para, (x,y), (x+w, y+h), (0,255,0), 4)
    cv2.putText(kontorlu_para, str("SK"), (x+w// 10-5,y+h // 5+3), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)
    cv2.putText(kontorlu_para, str(i+1), (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.9, (0,0,255), 3)

    print(cv2.contourArea(kontor))
plt.imshow(kontorlu_para)
plt.axis("off")
```
